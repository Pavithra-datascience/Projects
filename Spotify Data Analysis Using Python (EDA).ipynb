```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os

# Try to find the 'tracks.csv' file, first in the current directory
# then in the parent directory, and finally, prompting the user.
file_path = 'tracks.csv'  # First try in the current directory
if not os.path.exists(file_path):
    # If it doesn't exist, try the parent directory
    file_path = os.path.join('..', 'tracks.csv')

# Check if the file exists in the parent directory.
if not os.path.exists(file_path):
    # If it doesn't exist in either location, prompt the user for the path
    file_path = input("Please enter the correct path to 'tracks.csv': ")

# Now, check if the user-provided path exists
if not os.path.exists(file_path):
    raise FileNotFoundError(f"File not found at: {file_path}")

# Load your data into df_tracks
df_tracks = pd.read_csv(file_path)

# Create the sample dataframe
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))

# Now create the regression plot
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="b").set(title="Popularity vs Acousticness Correlation")
plt.show()
```

```python
df_tracks['dates'] = df_tracks.index.to_series()  # Access the datetime index directly
df_tracks.dates = pd.to_datetime(df_tracks.dates)  # This line might not be necessary
years = df_tracks.dates.dt.year  # Extract the years
```

```python
# Ensure the column 'duration' exists before trying to access it.
# If it doesn't, recalculate it from 'duration_ms'
if 'duration' not in df_tracks.columns:
    df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
    df_tracks.drop("duration_ms", inplace=True, axis=1)

total_dr = df_tracks.duration  # Now access the 'duration' column
sns.set_style(style="whitegrid")
# fig_dims was a single float, but figsize needs a tuple (width, height)
fig_dims = (10.5, 6)  # Example: 10.5 inches wide, 6 inches tall
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
import os
import pandas as pd
import zipfile

# Get the current working directory
current_directory = os.getcwd()
print(f"Current working directory: {current_directory}")

# Construct the full path to the ZIP file
zip_file_path = os.path.join(current_directory, 'archive.zip')

# Check if the ZIP file exists
if not os.path.exists(zip_file_path):
    print(f"Error: ZIP file '{zip_file_path}' not found.")
    zip_file_path = input("Please enter the correct path to the ZIP file: ")

# Specify the CSV file to read within the ZIP archive
csv_file_name = 'tracks.csv'  # Or 'artists.csv' if you want to read that file

# Read the CSV file from the ZIP archive
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    df_genre = pd.read_csv(zip_ref.open(csv_file_name))

# Now you can work with df_genre as usual
print(df_genre.head())
```
